*The following code makes a call to all the libraries needed for data processing in the assignment. This is also where I make a call to scrub data from a website.
I utilized the website www.city-data.com to pull a list of all the current zip codes in Austin, Texas*

```python
import requests
import numpy as np
import pandas as pd
import lxml
from bs4 import BeautifulSoup
import csv
!pip install uszipcode 
from uszipcode import SearchEngine

column_names = ['Zip Code']
zipcodedf = pd. DataFrame(columns = column_names)
zipcodes_df = pd.DataFrame(columns = {'Zip Code', 'Latitude', 'Longitude'})

web_link = 'http://www.city-data.com/city/Austin-Texas.html'

web_link_page = requests.get(web_link).text

web_link_raw = BeautifulSoup(web_link_page, 'lxml')

table_data = web_link_raw.find('div', class_ = 'container-fluid').find('section', class_ = 'zip-codes').text

search = SearchEngine(simple_zipcode=False)
i = 0


for zips in  table_data.split(', '):
    zipcodedf = zipcodedf.append({'Zip Code': zips},
                        ignore_index=True)



zipcodedf.loc[0,'Zip Code'] = zipcodedf.loc[0,'Zip Code'].split(' ')[2]
zipcodedf.loc[52,'Zip Code'] = zipcodedf.loc[52,'Zip Code'].split('.')[0]
#print(zipdf)

#print(zipcodedf)
```

```python


for zipcode in zipcodedf['Zip Code']:
    zipcoder = search.by_zipcode(zipcode)
    latitude = zipcoder.lat
    longitude = zipcoder.lng
    zipcodes_df = zipcodes_df.append({'Zip Code': zipcodedf.loc[i,'Zip Code'], 'Latitude' : latitude, 'Longitude' : longitude}, ignore_index=True)
    i = i+1
    
zipcodes_df
```

```python
from pandas.io.json import json_normalize
venues = results['response']['venues']

venue_data = json_normalize(venues)
venue_data
```

```python

filtered_columns = ['name', 'categories'] + [col for col in venue_data.columns if col.startswith('location.')] + ['id']
venuedata_filtered = venue_data.loc[:, filtered_columns]


def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']

venuedata_filtered['categories'] = venuedata_filtered.apply(get_category_type, axis=1)

venuedata_filtered.columns = [column.split('.')[-1] for column in venuedata_filtered.columns]

venuedata_filtered
```

```python
venues_grouped = venuedata_filtered.groupby(['postalCode'])
venues_count = venues_grouped.agg({'postalCode': 'count'})

venues_count
```
